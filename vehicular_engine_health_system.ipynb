# Vehicular Engine Health Monitoring System

This notebook contains the **complete Tkinter-based Python application** converted into a Jupyter Notebook format for documentation and reference.

```

# NOTE:
# This notebook contains the original Tkinter-based Vehicle Health Monitoring System code.
# For best results, copy this code into a .py file and run it using Python.

import numpy as np
import pandas as pd
import pyttsx3
import tkinter as tk
from tkinter import messagebox
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
import threading
import os

engine = pyttsx3.init()
engine.setProperty('rate', 150)

model = None
scaler = None
df = None

def load_and_train_model():
    global model, scaler, df
    df = pd.read_csv("New data set.csv")

    X = df[["GPS speed", "OBD speed", "Engine RPM", "Throttle position", 
            "Engine load", "Coolant Temperature"]]
    y = df[["Fuel Consumption", "CO2 Emission", "Engine Temperature", 
            "Brake Oil", "Maintenance Score"]]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42
    )

    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)

    model = MultiOutputRegressor(
        GradientBoostingRegressor(n_estimators=40, random_state=42)
    )
    model.fit(X_train, y_train)

print("Vehicle Health Monitoring System loaded.")
```
